```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
from sklearn import linear_model
```

```python
car= pd.read_csv('carprices.csv')
```

```python
car
```

```python
car = car.rename(columns={'Age(yrs)': 'Age'})
car = car.rename(columns={'Sell Price($)': 'Sell Price'})
```

```python
%matplotlib inline

plt.scatter(car['Mileage'], car['Sell Price'])
plt.show()
```

```python
plt.scatter(car['Age'], car['Sell Price'])
plt.show()
```

```python
#Since we can draw a straight line here, we will use linear regression model

x= car[['Mileage', 'Age']]
y= car['Sell Price']
```

```python
x
```

```python
y
```

```python
from sklearn.model_selection import train_test_split
```

```python
# Test_size= 0.2 means, 80% is training data and 20% is testing data
x_train, x_test, y_train, y_test = train_test_split(x,y, test_size= 0.2, random_state= 10)
```

```python
x
```

```python
len(car)
```

```python
len(x_train)
```

```python
len(x_test)
```

```python
reg= linear_model.LinearRegression()
```

```python
reg.fit(x_train, y_train)
```

```python
reg.predict(x_test)
```

```python
y_test
```

```python
# Checking accuracy of model, by calling score methodreg
reg.score(x_test, y_test)
```
